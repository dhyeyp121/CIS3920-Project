```python
import pandas as pd

df = pd.read_csv('data.csv', error_bad_lines=False)
```

```python
newdf = df.dropna(axis='columns')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
newdf.info()
```

```python
newdf.describe()
```

```python
newdf.describe(include='object')
```

```python
correlation_matrix = newdf.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
numerical_features = newdf.select_dtypes(include=['int64', 'float64']).columns
newdf[numerical_features].hist(bins=20, figsize=(12, 10))
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=newdf[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```
